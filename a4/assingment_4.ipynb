### (a)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')
```

```python
points = make_blobs(n_samples = 300,
                      n_features = 2,
                      centers = 4,
                      cluster_std = 0.6,
                      random_state=12345)
```

### (b)

```python
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import contingency_matrix

clusters = list()
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k,
                    random_state=None)
    kmeans.fit(points[0])
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    inertia = kmeans.inertia_
    clusters.append([labels, centers, inertia])
```

```python
for k in range(1, 11):
    plt.scatter(points[0][:,0], points[0][:,1], c = clusters[k-1][0])
    plt.scatter(clusters[k-1][1][:,0], clusters[k-1][1][:,1],
               c = "red", marker = "x", s = 60)
    plt.title("k = " + str(k))
    plt.show()
    cm = contingency_matrix(labels_true=points[1],
                            labels_pred=clusters[k-1][0])
    print("Contingency Matrix for k = " + str(k) + ":")
    display(pd.DataFrame(cm))
```

### (c)
ANSWER: Yes, the plot clearly shows that the natural number of clusters is 4, as there is the most substantial change of slope in the curve.

```python
sse = list()
for k in range(1, 11):
    sse.append(clusters[k-1][2])
print(sse)
plt.plot(range(1, 11), sse)
plt.show()
```

## Assignment 2
### (a)

```python
from scipy.cluster.hierarchy import dendrogram, linkage
vert = pd.read_csv("data/vertebrate.csv")
```

### (b)

```python
# preprocessing
names = list(vert["Name"])
vert = vert.set_index(["Name"])
classes = vert["Class"]
vert = vert.drop(["Class"], axis=1)
```

```python
# Single-link Hierarchical Clustering (MIN distance)
hc = linkage(vert, method="single")

plt.figure(figsize=(10, 5))
plt.title('Single-link Hierarchical Clustering Dendrogram')
plt.xlabel('Instance')
plt.ylabel('distance')
dendrogram(hc, labels=names,
           leaf_rotation=60)
plt.show()
```

```python
# Max-link Hierarchical Clustering (MAX distance)
hc = linkage(vert, method="complete")

plt.figure(figsize=(10, 5))
plt.title('Max-link Hierarchical Clustering Dendrogram')
plt.xlabel('Instance')
plt.ylabel('distance')
dendrogram(hc, labels=names,
           leaf_rotation=60)
plt.show()
```

```python
# Average-link Hierarchical Clustering
hc = linkage(vert, method="average")

plt.figure(figsize=(10, 5))
plt.title('Average-link Hierarchical Clustering Dendrogram')
plt.xlabel('Instance')
plt.ylabel('distance')
dendrogram(hc, labels=names,
           leaf_rotation=60)
plt.show()
```

Of these three hierarchical clustering methods, the maximum/complete linkage method provides the most natural representation.
